Посмотрим статистику игроков "Торонто Рэпторс" в их чемпионский сезон

```python
df.Team.unique()
```

```python
df = df.rename(columns={
  'Season Start Year':'Season_Start_Year',
  'Season Type':'Season_Type',
  'Player ID':'Player_ID',
  'Team ID':'Team_ID',
  'Games Played':'Games_Played',
  'Minutes Played':'Minutes_Played',
  'FG Made':'FG_Made',
  'FG Attempts':'FG_Attempts',
  'FG %':'FG_per',
  '3-Pt FG Made': 'FG_Made_3Pt',
  '3-Pt FG Attempts':'FG_Attempts_3Pt',
  '3-Pt FG %':'FG_per_3Pt',
  'FT Made':'FT_Made',
  'FT Attempts':'FT_Attempts',
  'FT %':'FT_per',
  'Offensive Rebounds':'Offensive_Rebounds',
  'Defensive Rebounds':'Defensive_Rebounds',
  'Personal Fouls':'Personal_Fouls',
  'Points Scored':'Points_Scored' 
})
```

Торонто взяли чемпионство в 2019 году, значит сезон начался в 2018

```python
team_tag = "TOR"
Toronto_champ_season = df.query('Season_Start_Year == 2018 & Team == @team_tag')
Toronto_champ_season.head()
```

Посмотрим статистику игроков по основным показателям в сезоне

```python
Toronto_champ_season.value_counts('Season_Type')
```

В регулярном сезоне за Торонто играло 18 человек

```python
Toronto_champ_season.info()
```

```python
Type_of_Season = 'Regular Season'
Tor_reg_stat = Toronto_champ_season.query('Season_Type == @Type_of_Season').iloc[:,[5, 8, 9, 21, 22, 23, 24, 25, 26, 27]]
Tor_reg_stat.info()
```

Введём несколько новых признаков, показывающих количество времени, проводимого на площадке в среднем, и эффективность в защите

```python
Tor_reg_stat['STL/FOUL'] = round(Tor_reg_stat.Steals / Tor_reg_stat.Personal_Fouls, 3)
Tor_reg_stat['Avg_minutes'] = round(Tor_reg_stat.Minutes_Played / Tor_reg_stat.Games_Played)
```
